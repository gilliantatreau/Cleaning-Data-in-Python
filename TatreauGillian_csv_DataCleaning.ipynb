## Step 1
Changed column name of '  num_pages' to strip the column name of white-spaces. 

```python
# change column name '  num_pages'
df.rename(columns = {'  num_pages': 'num_pages'}, inplace = True)
df.columns
```

## Step 2
Changed data types of "average_rating", "num_pages", and "isbn13" to numeric data types. These columns all contain numerical values and not string data. Changed "publication_date" to datetime data type, as it is the date of publication, not a string or numerical value. 

```python
# change data type of average_rating, num_pages, isbn13 to numeric
df["average_rating"] = pd.to_numeric(df["average_rating"], errors= "coerce")
df["num_pages"] = pd.to_numeric(df["num_pages"], errors= "coerce")
df["isbn13"] = pd.to_numeric(df["isbn13"], errors= "coerce")
```

```python
# change data type of publication_date to datetime
df["publication_date"] = pd.to_datetime(df["publication_date"])
df.dtypes
```

## Step 3
Looking at duplicate values. Changed "language_code" values to be consistent with ISO 639-2 standardized language codes.

```python
# number of unique values
df.nunique()
```

```python
# all unique values for "language_code"
print(df["language_code"].unique())
```

```python
# change "language_code" to be consistent with ISO 639-2 language codes
df["language_code"] = df["language_code"].replace(["nl", "fre", "en-US", 
                                                   "en-GB", "en-CA", "ger", 
                                                   "wel"], 
                                                  ["nld", "fra", "eng", 
                                                   "eng", "eng", "deu", 
                                                   "cym"])
print(df["language_code"].unique())
```

```python
# describe- looking for bad data, obvious outliers
df.describe()
```

```python
# making sure that the books with number of pages >= 2000 are boxed sets 
# or expected to be large books and not obvious errors
print(df[df["num_pages"] >= 2000])
```

## Step 4
Created new column "book_series" where "y" value denotes a book series and "n" value denotes a single book. Distinction made by title containing any of these key words: set, collection, novels, vols, volumes, books, ultimate.

```python
# adding column "book_series" where "y" denotes a book series and 
# "n" denotes a single book
df.loc[df["title"].str.lower().str.contains(
    "set|collection|novels|vols|books|volumes|ultimate"),'book_series'] = 'y'
df["book_series"].fillna("n", inplace = True)
```

## Step 5
Changed all strings in DataFrame to lowercase only.

```python
# make all strings lowercase
df = df.applymap(lambda s: s.lower() if type(s) == str else s)
```

## Step 6
Separated "authors" column into "author" (the first author listed), and "contributors" (the rest of the authors listed in authors). Deleted original "authors" column. 

```python
# separate "authors" into primary author (first listed in authors list)
# and contributors (rest of authors list)
new = df["authors"].str.split("/", n = 1, expand = True)
df["author"] = new[0]
df["contributors"] = new[1]
df.drop(columns = ["authors"], inplace = True)
```

## Step 7
Rearranged column order to make the most sense reading it- "bookID" and "isbn" are first as those are both unique identifiers for the entries; then "title", "author", "contributors", "book_series" (information that seems like it would possibly be useful/common across all three data sets); then the numerical columns; and finally "isbn13", "language_code", "publication_date", and "publisher" as these do not intuitively feel as informative from a data analysis stand point but should not be removed at this early of a stage. 

```python
# rearrange dataframe column order
df = df.iloc[:,[0,3,1,12,13,11,2,6,7,8,4,5,9,10]]
```

## Ethical Implications

Some of the ethical implications considered while cleaning this data set include: changing the language codes, how book_series column was determined, separating the "authors" column, and the decision to not remove duplicates of the same title. The original "language_code" data included en_US, en_GB, and en_CA for American, British, and Canadian english, respectively. While all these are at the end of the day, in English, there are some differences in slang usage and spelling between the respective versions. By removing the distinction between them, there may be added confusion between versions, but the ISBNs will still be unique and different, so version differentiation is still possible only slightly more complicated. The "book_series" column was created by only using a list of some key words, but it is more than likely that not all book series contained any of those words and that some titles of singular books might have been included on accident, therefore, the book_series column is really only, at this point, an estimation and would require parsing through the entire data set to validate and correct it as accurately as possible. The separation of the "authors" column was done to simplify the data and make it more readable; however, it is possible that this separation was not done in a way that honors all contributors to the level they are due. By assuming the first author listed was the primary author lessons the impact if there are authors of equal importance and it also assumes that the list was written in the order of primary author first. The last ethical implication from this data set was one considered but not used: there are duplicates of book titles in the data set, but none were removed because there were no duplicates in ISBN. This means that multiple versions (audiobook, ebook, hardcover, paperback) and editions of several titles are inlcuded on this list. Since each ISBN has its own, average_rating, ratings_count, and text_reviews_count, it did not seem ethical to combine these values in some way to account for duplicates or to simply remove them entirely as each title drew its own demographic of readers and reviewers.
